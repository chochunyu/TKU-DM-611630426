```python
import matplotlib.pyplot as plt
import numpy as np 
from sklearn import datasets,linear_model
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes_X = diabetes.data[:,np.newaxis,2]
```

```python
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]
```

```python
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(diabetes_X_train,diabetes_y_train)
```

```python
print('Coefficients:\n',regr.coef_)
```

```python
print('Mean squared error: %.2f' % 
      np.mean((regr.predict(diabetes_X_test) - diabetes_y_test)**2))
```

```python
print('Variance score: %.2f' % regr.score(diabetes_X_test, diabetes_y_test))
```

```python
plt.scatter(diabetes_X_test,diabetes_y_test, color = 'black')
plt.plot(diabetes_X_test, regr.predict(diabetes_X_test),color = 'blue',linewidth=3)

plt.xticks(())
plt.yticks(())
```

```python
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from time import time
import numpy as np
import matplotlib.pyplot as plt

from sklearn import metrics
```

```python
X= np.array([[1,1],[1.1,1.1],[1.2,1.2],
             [2,2],[2.1,2.1],[2.2,2.2]])
y=[1,1,1,
  0,0,0]
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print("集群中心的座標:",kmeans.cluster_centers_)
print("預測:",kmeans.predict(X))
print("實際:",y)
print("預測[1, 1],[2.3,2.1]:",kmeans.predict([[1, 1],[2.3,2.1]]))
```

```python
plt.axis([0, 3,0, 3])
plt.plot(X[:3,0],X[:3,1],'yx')
plt.plot(X[:3,0],X[:3,1],'g.')
plt.plot(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], 'ro')
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score

df = pd.read_csv('customer.csv')
dict1 = {'男':1, '女':2}
df['性別'].replace(dict1, inplace=True)
df
```

```python
km = KMeans(n_clusters=3)
km.fit(df)
km.labels_
```

```python
df['類別'] = km.labels_
df.head()
```

```python
df2 = df[df['類別']==0]
df3 = df2.iloc[0:30, :]
df3
```

```python
df2 = df[df['類別']==1]
df3 = df2.iloc[0:30, :]
df3
```

```python
df2 = df[df['類別']==2]
df3 = df2.iloc[0:30, :]
df3
```

```python
for n in range(2,15): 
  km = KMeans(n_clusters=n)
  km.fit(df)
  metric = calinski_harabasz_score(df, km.labels_) 
  print('群組數量：{}，評分：{}'.format(n, metric))
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('customer.csv')
dict1 = {'男':1, '女':2}
df['性別'].replace(dict1, inplace=True)
df
```
